## **Matplotlib**.

Es una biblioteca para crear visualizaciones estáticas, animadas e interactivas utilizando Python. Entre sus características se tiene:

* Crea gráficas de calidad.
* Las gráficas son interactivas.
* El estilo visual puede personalizarse.
* Se pueden exportar a muchos formatos.
* Se integran aJupyter Lab e interfaces de usuario.
* Utiliza arreglos ricos o de terceros.

En su [sitio oficial](https://matplotlib.org/), es posible encontrar diversos recursos como docmentación, ejemplos, tutoriales, etc.

Para instalar matplotlib es necesario hacerlo con el administrador de paquetes `pip`:

`pip install matplotlib`

O bien utilizando `conda`:
`conda install matplotlib`

Es común utlizar arreglos *numpy* y combinarlos con las capacidades de graficación de Matplolib.

Crear una gráfica de líneas  con títulos y descripción de los ejer X y Y.

```python
import matplotlib.pyplot as plt #Para gráficas
import numpy as np #Cálculos numéricos y arreglos

x = np.arange(-5.0, 5.0, 0.01) #Arreglo uno...desde el -0.5 hasta el 5.0 con espaciado de 0.01(resolución)
y = np.sin(np.pi*x) #Seno (Calcular el valor de seno, multiplicando pi * x)

#Graficar
plt.plot(x, y) #Dibujar linea continua conectando los puntos x,y
plt.title("Ejemplo de gráfica") #Titulo del grafico
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show() #Mostrar la gráfica
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-5.0, 5.0, 0.01) #Arreglo uno...desde el -0.5 hasta el 5.0 con espaciado de 0.01(resolución)
y = np.tan(np.pi*x) #Tangente (tan)

#Graficar
plt.plot(x, y)
plt.title("Ejemplo de gráfica") #Titulo del grafico
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show() #Mostrar la gráfica
```

Se pueden personalizar los indicadores de los valores a graficar, el color y tipo de la línea que se dibujará. Los valores aceptados para el tipo de línea son: `'-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'`.

```python
x = np.arange(20)
y = np.random.randint(1, 100, 20) #Llenado aleatorio enteros del 1-100

plt.plot(x, y, marker = "o", color = "orange", ls= "dotted" ) #Marcar con un circulo negro donde cambia el gráfico
plt.title("Ejemplo")
plt.xlabel("Eje X")
plt.ylabel ("Eje Y")
plt.show()
```

***Pandas*** es una bibilioteca que es utilizada para trabajar con análisis y manipulaciónde datos. comúnmente se utiliza en combinación con Matplotlib para mostrar dichos datos de manera gŕafica.
El objeto base para trabajar con *Pandas* es el *DataFrame*, que esencialmente es una estructura de datos bidimensional, similar a un arreglo o a una tabla con filas y columnas.
Par amayor información sobre *Pandas", se puede consultar su [sitio oficial](https://pandas.pydata.org/).
En algunos ejemplos posteriores, se utilizará *Pandas* y *numpy* para suministrar datos a *Matplotlib*.

```python
#Para analisis de datos
import matplotlib.pyplot as plt
import pandas as pd #Para trabajar con estructuras de datos tipo tabla (DataFrame).
import numpy as np

n = 100 #Generara 100 registros
columns = ["Marca"+str(number) for number in range(1, 5)] #Crear un arreglo (Marca 1, Marca 2...)
date = pd.date_range(start="2021-1-1", periods = n) #Indice de fechas, 100 días consecutivos a partir del 1 de enero de 2021.
data = np.random.randint(low = 1, high=[100, 300, 45, 77], size=(n, 4))
df = pd.DataFrame(data, columns = columns, index = date) #Crear una tabla

plt.figure(figsize = (12, 5)) #Crear figura (ancho, alto)
plt.plot(df.index, df["Marca1"], c = "red", ls="-") #Grafica los valores de Marca1 a lo largo del tiempo.
plt.title("Marca A")
plt.xlabel("Fecha")
plt.ylabel("Valores")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

n = 100
columns = ["Marca"+str(number) for number in range(1, 5)] #Crear un arreglo (Marca 1, marca 2...)
date = pd.date_range(start="2021-1-1", periods = n)
data = np.random.randint(low = 1, high=[100, 300, 45, 77], size=(n, 4))
df = pd.DataFrame(data, columns = columns, index = date)

plt.figure(figsize = (12, 5))
plt.plot(df.index, df["Marca2"], c = "blue", ls="-")
plt.title("Marca A")
plt.xlabel("Fecha")
plt.ylabel("Valores")
plt.grid(True)
plt.show()
```

Con la función `plt.bar()` es posible crear gráficas de barras:



```python
data = df.sum().sort_values(ascending=False) #Suma los valores y los ordenada de mayor a menor en forma descendente

plt.bar(x = data.index, height=data)
plt.xlabel("Marcas")
plt.ylabel("Valores")
plt.title("Ejemplo de gráfica de barras")
plt.show()
```

Las gráficas de barras pueden ser también horizontales.

El método `savefig()` se usa para guardar en el disco duro la gráfica creada en formato de imagen.

```python
plt.barh(y = data.index, width=data)
plt.xlabel("Marcas")
plt.ylabel("Valores")
plt.title("Ejemplo de gráfica de barras")
plt.savefig("EjemploHorizontal.jpeg") #Savefig() para guardar en el disco local
plt.show()
```

Otra biblioteca que se usa comúnmente en combinación con *matplotlib* es *scikit-learn*. Está última esencialmente sirve para utiliar algoritmos de aprendizaje máquina de una manera realtivamente sencilla. Más información en su [sitio oficial](https://scikit-learn.org/stable/index.html).

En el ejericio siguiente, se va a utilizar para dibujar un diagrama de dispersión con los marcadores indicados con el símbolo "^" y una opacidad de 0.3.

```python
from sklearn.datasets import make_regression
#import matplotlib.pyplot as plt

x, y = make_regression(n_samples=100, n_features=1, noise=20)

plt.scatter(x, y, color="red", alpha=0.3) #alpha para opacidad
plt.title("Relación entre X y Y")
plt.xlabel("Características")
plt.ylabel("Objetivo")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

Para construir un histograma se usa la función `plt.hist()`.

```python
data = np.random.normal(loc=100, scale=0.5, size=100)

plt.hist(x = data) #Para histogramas
plt.title("Histograma")
```

En la gráfica siguiente se cambiará el tamaño de la fuente a 15 en los ejes y a 20 en el título. La función `plt.legend()` permitirá insertar una leyenda en el gráfico y con el argumento `loc`, se indicará la posición. Las opciones válidas para dicho argumento son: `best, 	upper right, upper left, lower left, lower right, right, center left, center right, lower center, upper center, center`.

```python
x = np.arange(1, 11)
y1 = x**2
y2 = x**3
y3 = x**4

plt.plot(y1, 'o', y2, '^',y3, "bs")
plt.legend(["y1", "y2", "y3"], loc="upper left") #Agregar una leyenda y en que posición
plt.title("Valores", fontsize=20)
plt.xlabel("x", fontsize=15)
plt.ylabel("y", fontsize=10)
plt.grid(linestyle="--", alpha=0.8, color="red")
plt.show()
```

Para crear una gráfica de barras apiladas, es necesario agregar en el mismo eje, los conjuntos de datos que se quieran mostrar.

```python
marca_A = [120, 130, 145, 177, 270, 211]
marca_B = [90, 41, 140, 150, 230, 193]
meses = ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio"]

plt.bar(x = meses, height=marca_A, color="red", label="Marca A")
plt.bar(x = meses, height=marca_B, color="blue", label="Marca B")
plt.title("Ingresos sobre el tiempo")
plt.show()
```

Para una gráfica de pastel, se tiene que usar la función `plt.pie()`:

```python
ventas = [1500.2, 1378, 671, 431, 700]
etiquetas = ["Marcas"+str(number) for number in range(1, len(ventas)+1)]

plt.pie(ventas, labels = etiquetas) #Gráfica de pastel
plt.show()
```

También se le puede agregar sombra, una leyenda, los pocentajes correspondientes y separar cada segmento:

```python
separacion = (0.1, 0.1, 0, 0, 0.1)

plt.pie(ventas, labels = etiquetas, autopct='%1.2f%%', explode=separacion, shadow=True)
plt.legend(etiquetas, loc = "best")
plt.show()
```

En el ejercicio siguiente, se va a crear un dataframe con valores al azar, se imprimirá la tabla y la gráfica correspondiente. Los valores cambiarán respecto al tiempo (cada mes). Cada columna será representada con un línea de color diferente.

```python
np.random.seed(7)

data = np.random.randint(low=100, high=[250, 300, 500, 1000], size=(12, 4))
index = pd.date_range(start="2021-1-1", periods=12, freq="M")
df = pd.DataFrame(data, columns=list("ABCD"), index=index.month_name())

print(df)

plt.plot(df)
plt.title("Valores sobre el tiempo")
plt.legend(df.columns, loc="best")
plt.grid(linestyle="--", alpha=0.5)
plt.xlabel("Meses")
plt.ylabel("Ventas")
plt.show()
```

Como se observa en el resultado anterior, la gráfica se nota un tanto ilegible dado el tamaño de la misma, es posible modificar lo anterior con el argumento `figsize`. Por otra parte, si queremos modificar el límite del eje Y, se hace con la función `plt.ylimit()`. Modificar estos parámetros, permitirá generar una mejor gráfica.

```python
plt.figure(figsize=(15, 4))
plt.plot(df)
plt.title("Valores sobre el tiempo")
plt.legend(df.columns, loc="best")
plt.grid(linestyle="--", alpha=0.5)
plt.xlabel("Meses")
plt.ylabel("Ventas")
plt.ylim(80, 1000)
plt.show()
```
